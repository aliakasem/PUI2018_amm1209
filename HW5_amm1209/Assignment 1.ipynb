```python
query = '''SELECT LandUse, COUNT(ZipCode) as ZipCode FROM manhattan WHERE LandUse IS NOT NULL GROUP BY LandUse; '''
manhattan_grouped = pdsql.sqldf(query, locals())
```

```python
manhattan_grouped
```

```python
query = '''SELECT LandUse, COUNT(ZipCode) as ZipCode FROM brooklyn WHERE LandUse IS NOT NULL GROUP BY LandUse; '''
brooklyn_grouped = pdsql.sqldf(query, locals())
```

```python
brooklyn_grouped
```

```python
manhattan_grouped['LandUse'] = manhattan_grouped['LandUse'].astype(int)
```

```python
manhattan_grouped
```

```python
brooklyn_grouped['LandUse'] = brooklyn_grouped['LandUse'].astype(int)
```

```python
brooklyn_grouped
```

```python
man_bk = pd.merge(manhattan_grouped,brooklyn_grouped, on = 'LandUse')
```

```python
man_bk
```

```python
man_bk.columns
```

```python
man_bk = man_bk.rename(index=str, columns={"ZipCode_x": "Manhattan", "ZipCode_y": "Brooklyn"})
```

```python
man_bk
```

```python
LandUse_dic ={1:"One & Two Family Buildings",2:"Multi-Family Walk-Up Buildings",3:"Multi-Family Elevator Buildings", 4:"Mixed Residential & Commercial Buildings", 5: "Commercial & Office Buildings", 6:"Industrial & Manufacturing", 7:"Transportation & Utility", 8: "Public Facilities & Institutions", 9:"Open Space & Outdoor Recreation",10:"Parking Facilities",11:"Vacant Land"} 

man_bk = man_bk.replace({"LandUse": LandUse_dic})
```

```python
man_bk
```

```python
man_bk['land_use_ratio'] = man_bk.Manhattan/man_bk.Brooklyn
```

```python
man_bk
```

```python
def wins(x):
    if x>1:
        return True
    else:
        return False
man_bk['Manhattan wins'] = man_bk['land_use_ratio'].apply(wins)
```

```python
man_bk
```

From the last table, we can see that Manhattan only wins in amount of buildings in the Multi-Family Elevator Buildings, which makes sense, since Manhattan is much smaller than Brooklyn and has taller buildings as well.
